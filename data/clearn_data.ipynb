```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
IMU_Data = '/home/motaz/Projects/graduation_project_good/2nd_GP_Hardware/hardware/graduation_hardware/imu/robot_data_2025-11-20_22-06-35.csv'

df = pd.read_csv(IMU_Data)


print(df.head())
```

```python
# Define the columns that need to check for duplicates (Timestamp Ignored)
subset_cols = [ 'Kappa', 'Phi', 'X', 'Y', 'Z']

# Drop duplicates on these columns
# 'first' argument is for keep the first occurrence 
cleaned_df = df.drop_duplicates(subset=subset_cols, keep = 'first')
```

```python
print(f" Original data rows : {len(df)}")

print (f" Cleaned rows : {len(cleaned_df)}")
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)
ax1.plot(cleaned_df['Timestamp'], cleaned_df['X'], label='X')
ax1.plot(cleaned_df['Timestamp'], cleaned_df['Y'],label= 'Y')
ax1.plot(cleaned_df['Timestamp'],cleaned_df['Z'], label = 'Z')


ax2.plot(cleaned_df['Timestamp'], cleaned_df['Kappa'], label = 'Kappa')
ax2.plot(cleaned_df['Timestamp'], cleaned_df['Phi'], label = 'phi')

ax3.plot(cleaned_df['Timestamp'], cleaned_df['Theta1'], label = 'Theta1')
ax3.plot(cleaned_df['Timestamp'], cleaned_df['Theta2'], label = 'Theta2')
ax3.plot(cleaned_df['Timestamp'], cleaned_df['Theta3'], label = 'Theta3')


ax1.legend()
ax2.legend()
ax3.legend()
```
